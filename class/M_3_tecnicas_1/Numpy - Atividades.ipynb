```python
from scipy.io.wavfile import read
import matplotlib.pyplot as plt
```

```python
# read audio samples
input_data = read("Guns-N-Roses-Estranged.wav")
input_data
```

```python
audio = input_data[1]
print(audio.shape)
type(audio)
```

```python
# plot the first 1024 samples
plt.plot(audio[0:44100*10])
# label the axes
plt.ylabel("Amplitude")
plt.xlabel("Time")
# set the title  
plt.title("Sample Wav")
# display the plot
plt.show()
```
