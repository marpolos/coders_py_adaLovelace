```python
df_info = pd.DataFrame({
    'id': [2, 3, 4, 5],
    'age': [31, 20, 40, 70],
    'sex': ['F', 'M', 'M', 'F']
})
df_info
```

```python
df_all_data = df_customer.merge(df_info, how='inner', on='id')
df_all_data
```
