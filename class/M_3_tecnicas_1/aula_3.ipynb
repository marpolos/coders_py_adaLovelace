```python
# Funções notáveis

import numpy as np

#where

a = np.arange(10)
a
```

```python
np.where(a>5, 10, 0)
```

```python
b = np.where(a>5, a, 0)
```

```python
#unique

np.unique(b)
```

```python
valores, freq = np.unique(b, return_counts= True)
print(f'Valores: {valores} frequencia: {freq}')
```

```python
#argmax
index_max = np.argmax(freq)
index_max
```

```python
#moda
moda = valores[index_max]
moda
```
