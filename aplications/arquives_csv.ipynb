```python
with open("users.csv", "w", newline="") as file:
    escritor = csv.writer(file)
    escritor.writerow(["nome", "sobrenome"])
    escritor.writerow(["Marcelle", "Monteiro"])
    escritor.writerow(["Jhonata", "Goulart"])
    escritor.writerow(["Bernardo", "Geraldo"])
```

```python
header = ["nome", "sobrenome"]
dados = []

opt = input("O que deseja: 1 - cadastrar ou 0 - sair ?")
while opt != "0":
    nome = input("Nome: ")
    sobrenome = input("Sobrenome: ")
    dados.append([nome, sobrenome])
    opt = input("O que deseja: 1 - cadastrar ou 0 - sair ?")

with open("usuarios.csv", "w") as file:
    writer = csv.writer(file)
    writer.writerow(header)
    writer.writerow(dados)

with open("usuarios.csv", "r") as file:
    reader = csv.reader(file, delimiter=",")
    print(reader)


```
